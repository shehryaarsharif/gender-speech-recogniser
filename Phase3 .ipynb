# File Reading & Feature Extraction

```
def get_MFCC(audio, sr):
    features = mfcc.mfcc(audio, sr, 0.025, 0.01, 13, appendEnergy = True)
    return np.mean(features, axis=0)
```

```
Gender_Recognition_path='/content/dataset_project/Gender_Recognition/'
Speaker_Recognition_path='/content/dataset_project/Speaker_Recognition/'

Train_data_GR=[]
Test_data_GR=[]


Train_labels_GR=[]
Test_labels_GR=[]

Train_data_SR=[]
Test_data_SR=[]

Train_labels_SR=[]
Test_labels_SR=[]

Data_types_GR = os.listdir(Gender_Recognition_path)
Data_types_SR = os.listdir(Speaker_Recognition_path)


train_folder_GR=os.listdir(Gender_Recognition_path+"/"+Data_types_GR[1])
test_folder_GR=os.listdir(Gender_Recognition_path+"/"+Data_types_GR[0])

train_folder_SR=os.listdir(Speaker_Recognition_path+"/"+Data_types_SR[1])
test_folder_SR=os.listdir(Speaker_Recognition_path+"/"+Data_types_SR[0])



#training features for Gender Recognition
for folder in train_folder_GR:
  # print(folder[-1])
  if folder[-1] =="M":
    [Train_labels_GR.append(0) for i in range(10)]
  else:
    [Train_labels_GR.append(1) for i in range(10)]
  
  vn_folder=os.listdir(Gender_Recognition_path+"/"+Data_types_GR[1]+"/"+folder)
  for vn in vn_folder:
    sr, audio = read(Gender_Recognition_path+"/"+Data_types_GR[1]+"/"+folder+"/"+vn)
    features = get_MFCC(audio, sr)
    # print (features.shape)
    Train_data_GR.append(features)

#testing features for Gender Recognition
for folder in test_folder_GR:
  if folder[-1] =="M":
    [Test_labels_GR.append(0) for i in range(10)]
  else:
    [Test_labels_GR.append(1) for i in range(10)]
    
  vn_folder=os.listdir(Gender_Recognition_path+"/"+Data_types_GR[0]+"/"+folder)
  for vn in vn_folder:
    sr, audio = read(Gender_Recognition_path+"/"+Data_types_GR[0]+"/"+folder+"/"+vn)
    features = get_MFCC(audio, sr)
    # print (features.shape)
    Test_data_GR.append(features)




#training features for Speaker Recognition
for folder in train_folder_SR:
  [Train_labels_SR.append(int(folder[3:6]) - 1) for i in range(8)]
  vn_folder=os.listdir(Speaker_Recognition_path+"/"+Data_types_SR[1]+"/"+folder)
  for vn in vn_folder:
    sr, audio = read(Speaker_Recognition_path+"/"+Data_types_SR[1]+"/"+folder+"/"+vn)
    features = get_MFCC(audio, sr)
    # print (features.shape)
    Train_data_SR.append(features)

#testing features for Speaker Recognition
for folder in test_folder_SR:
  [Test_labels_SR.append(int(folder[3:6]) - 1) for i in range(2)]
  vn_folder=os.listdir(Speaker_Recognition_path+"/"+Data_types_SR[0]+"/"+folder)
  for vn in vn_folder:
    sr, audio = read(Speaker_Recognition_path+"/"+Data_types_SR[0]+"/"+folder+"/"+vn)
    features = get_MFCC(audio, sr)
    # print (features.shape)
    Test_data_SR.append(features)



Train_data_GR=np.array(Train_data_GR)
Train_data_SR=np.array(Train_data_SR)

Test_data_GR=np.array(Test_data_GR)
Test_data_SR=np.array(Test_data_SR)



Train_labels_GR=np.array(Train_labels_GR)
Train_labels_SR=np.array(Train_labels_SR)

Test_labels_GR=np.array(Test_labels_GR)
Test_labels_SR=np.array(Test_labels_SR)
```

# Part 1

## Gender Recognition

### Hyper-parameter Tuning Using Grid Search

```
# parameters = {'solver': ['sgd'], 'max_iter': [5000], "activation": ["logistic"], 'hidden_layer_sizes': [(128,64), (64),(64,32), (32)], 'random_state':[1], "learning_rate_init": [0.1, 0.4, 0.01]}


MLP = MLPClassifier(random_state=1, max_iter=5000, activation="logistic", solver="sgd")
tuned_params={"learning_rate_init": [0.4, 0.1, 0.01], 'hidden_layer_sizes': [(128,64),(64,1),(64,32),(32,1)]}

clf_GR = GridSearchCV(MLP, tuned_params, scoring="f1_macro",verbose=1)
clf_GR.fit(Train_data_GR,Train_labels_GR)
print(clf_GR.best_params_)
```

```
predictions=clf_GR.predict(Test_data_GR)
```

### Evaluation

```
print(classification_report(Test_labels_GR,predictions))
print(accuracy_score(Test_labels_GR,predictions))
print(confusion_matrix(Test_labels_GR,predictions))
```

Additional Evaluation

```
print("82% f1-macro achieved at LR = 0.005 and Hidden_Layers = (128, 64)")
```

## Part 2
Linear SVM

```
def LinearSVM(trainX, testX, trainLabels):
  LSV = svm.LinearSVC(random_state=1, max_iter=5000, dual=False)
  LSV.fit(trainX, trainLabels)
  return LSV.predict(testX)

LSV_predictions_GR = LinearSVM(Train_data_GR, Test_data_GR, Train_labels_GR)
LSV_predictions_SR = LinearSVM(Train_data_SR, Test_data_SR, Train_labels_SR)
```

### Gender Recognition

```
print("Accuracy on test GR:",accuracy_score(Test_labels_GR, LSV_predictions_GR))
print("================= {} ==================".format("GR Report on test data"))
print(classification_report(Test_labels_GR, LSV_predictions_GR))
print("================= {} ==================".format("Confusion Matrix"))
print(confusion_matrix(Test_labels_GR, LSV_predictions_GR))
```

### Speaker Recognition

```
print("Accuracy on test SR:",accuracy_score(Test_labels_SR, LSV_predictions_SR))
print("================= {} ==================".format("SR Report on test data"))
print(classification_report(Test_labels_SR, LSV_predictions_SR))
print("================= {} ==================".format("Confusion Matrix"))
print(confusion_matrix(Test_labels_SR, LSV_predictions_SR))
```

## Part 3
Naive Bayes

```
def NaiveBayes(trainX, testX, trainLabels):
  NB = GaussianNB()
  NB.fit(trainX, trainLabels)
  return NB.predict(testX)  

NB_predictions_GR = NaiveBayes(Train_data_GR, Test_data_GR, Train_labels_GR)
NB_predictions_SR = NaiveBayes(Train_data_SR, Test_data_SR, Train_labels_SR)
```
